### 1. Create

#### 1.1 From CSV

```python
df = pd.read_csv('sampledata.csv')
```

#### 1.2 From Dictionary

```python
mydict = {'col1':[1,2,3], 'col2':[4,5,6], 'col3':[7,8,9]}
```

### 2. Read

#### 2.1 Show Top 5 and Bottom 5 Rows

```python
# OUTPUT: CSV-looking table
# Can take an arg for rows wanted, default is 5
df.head()
```

```python
dictdf.tail(2)
```

#### 2.2 Show Columns and Data Types

```python
# OUTPUT: Index(['ColName1', 'ColName2', etc...], dtype='...')
df.columns
```

```python
# OUTPUT: Two column list - ColName, dtype
df.dtypes
```

```python
# OUTPUT:
"""
<class ___>
RangeIndex: _ entries, _ to _
Data columns (total _ columns)

Table of cols - #, Column, Non-Null Count, Dtype

dtypes: type(number of cols using it), nexttype(), etc...
memory usage: ___ KB
"""

df.info()
```

#### 2.3 Summary Statistics

```python
# OUTPUT: Table of statistics for each column
# [count, mean, std, min, 25%, 50%, 75%, max]

# Only will run stats on floats/ints
df.describe()
```

```python
# OUTPUT: Table of statistics for each column
# [count, unique, top, freq]

# Only runs stats on cols with dtype of object
df.describe(include='object')
```

#### 2.4 Filtering Columns

```python
# OUTPUT: Table of index, valueOfInputtedCol

# To get a single column
df['International plan']
```

```python
# OUTPUT: Table of index, valuecol1, valuecol2, ...

# To get multiple columns
df[['State', 'International plan']]
```

```python
# OUTPUT: Index(['ColName1', 'ColName2', etc...], dtype='...')
# Returns array of unique values

df['State'].unique()
```

#### 2.5 Filtering on Rows

```python
# OUTPUT: CSV-esque table, similar to head, 
#         but only w/ rows that fit the filters

# df[  df[colName]    condition    value ]
df[df['International plan'] == 'No']
```

```python
# To filter on multiple rows w/ multiple conditions
# Needs to use '&' for bitwise (True/False) operations

df[(df['International plan'] == 'No') & (df['Churn'] == False)]

# Shorter way to do it w/ query
#df.query("`International plan` == 'No' & Churn == False")
```

#### 2.6 Indexing with iloc
Indexing with integers (as opposed to specific values)

```python
# OUTPUT: Table of colName, colValue

df.iloc[10]
```

```python
# OUTPUT: Single value

# To get only a specific col: df[rowNum, colNum]
# You can also use negative indexing, -1 is last col etc.
df.iloc[14,0]
```

```python
# OUTPUT: Table of rows in range

# To slice [inclusive:exclusive]
df.iloc[22:24]
```

#### 2.7 Indexing with loc
Indexing with keywords
- Needs to have an index set

```python
# Set index on State
state = df.copy()
state.set_index('State', inplace=True)
```

```python
# OUTPUT: Full table of all rows whose value 
#         for the indexed row matches the provided val

state.loc['OR']
```

### 3.Update

#### 3.1 Dropping Rows

```python
# OUTPUT: Table of ColumnName, numRowsNull

# Count how many rows have null values for each column
df.isnull().sum()
```

```python
# Drop all rows with missing values
df.dropna(inplace=True)
```

#### 3.2 Dropping Columns

```python
# OUTPUT: Full dataset w/ dropped col removed

# Drop column by name (axis 0 = rows, axis 1 = columns)
df.drop('Area code', axis=1)
```

#### 3.3 Creating Calculated Columns

```python
df['New Column'] = df['Total intl minutes'] + df['Total night minutes']

# Shorter way using eval
df['New Column'] = df.eval('`Total intl minutes` + `Total night minutes`')
```

#### 3.4 Updating an Entire Column

```python
df['New Column'] = 100
```

#### 3.5 Updating a Single Value

```python
df.iloc[0, -1] = 10
```

#### 3.6 Condition Based Updating (using Apply)

```python
df['Churn Binary'] = df['Churn'].apply(lambda x: 1 if x else 0)
```

```python
# OUTPUT: First 5 rows where condition is true

df[df['Churn']==True].head()
```
